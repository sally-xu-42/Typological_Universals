```python
import stanza
import spacy_stanza
from spacy import displacy

print("INFO: Downloading English model...")
stanza.download("en", verbose=False)

print("INFO: Initialize the pipeline")
nlp = spacy_stanza.load_pipeline("en", processors='tokenize,mwt,pos,lemma,depparse', verbose=False, use_gpu=False)
```

```python
doc = nlp("I opened the door.")
displacy.render(doc, style="dep")
```

```python
doc = nlp("He is honest.")
displacy.render(doc, style="dep")
```

```python
doc = nlp("He seems honest.")
displacy.render(doc, style="dep")
```
